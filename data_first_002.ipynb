## 마스터테이블 변경
호텔 테이블에 있는 모든 호텔에 추천 후보 호텔을 연결한 데이터 작성

1.  big_area_name	small_area_name, hotel_cnt 계산
2.  join_area_id 를 만든다
3.  hotel_cnt > 20 이면 small_area_name 가져오기
4.  hotel_cnt < 20 이면 big_area_name 가져오기

```python
#불필요한 메모리 제거를 위한 라이브러리 불러오기
import gc
```

```python
# small_area_nam 호텔 계산
small_area_mst = hotel_tb.groupby(['big_area_name','small_area_name'],as_index=False).size()
small_area_mst
```

```python
small_area_mst.columns=['big_area_name','small_area_name','hotel_cnt']
small_area_mst
```

```python
# hotel_cnt 20건 이상이면 join_area_id를 small_area_name 가져오기
# hotel_cnt 20건 이하이면 join_area_id를 big_area_name 가져오기
# -1로 자신은 제외

small_area_mst['join_area_id']=np.where(small_area_mst['hotel_cnt']-1>=20, small_area_mst['small_area_name'],small_area_mst['big_area_name'])
small_area_mst
```

```python
#필요 없는 열 제거
small_area_mst.drop(['hotel_cnt','big_area_name'],axis=1, inplace=True)
small_area_mst
```

```python
#추천 전 호텔에 small_area_mst를 결합해 join_area_id를 설정
base_hotel_mst=pd.merge(hotel_tb,small_area_mst,on='small_area_name').loc[:,['hotel_id','join_area_id']]
base_hotel_mst
```

```python
# 아래는 필요에 따라 메모리를 해제한다
del small_area_mst
gc.collect()
```

```python
#recommend_hotel_mst 는  추천 후보 테이블
recommend_hotel_mst = pd.concat([
                                 hotel_tb[['small_area_name','hotel_id']].rename(columns={'small_area_name':'join_area_id'},inplace=False),
                                 hotel_tb[['big_area_name','hotel_id']].rename(columns={'big_area_name':'join_area_id'},inplace=False)])
recommend_hotel_mst
```

```python
#결합할때 hotel_id열 이름이 중복 됨
# 변경
recommend_hotel_mst.rename(columns={'hotel_id':'rec_hotel_id'},inplace=True)
recommend_hotel_mst
```

```python
#base_hoel_mst와 recommend_hotel_mst 결합하여 추천 후보의 정보를 부여 한다
# query 함수로 추천 후보에서 자신을 제외 한다

pd.merge(base_hotel_mst,recommend_hotel_mst,on='join_area_id').loc[:,['hotel_id','rec_hotel_id']].query('hotel_id != rec_hotel_id')
```
