인덱스를 이용한 행 추출

 자의적 샘플링 , 랜덤 샘플링, 공평한 샘플링

```python
#부작위  50% 샘플링
reserve_tb.sample(frac=0.5)
```

공평한 샘플링?
= 어느 한쪽에 편향 되지 않는 샘플링

```python
#reserve_tb['customer_id'].uniue() 중복을 제거한 customer_id를 반환
#smaple 함수로 고객 ID를 샘플링 한다.

target=pd.Series(reserve_tb['customer_id'].unique()).sample(frac=0.5)

#isin함수를 이용하여 customer_id가 샘플링한 고객 ID중에 존재하면 해당 행을 추출
reserve_tb[reserve_tb['customer_id'].isin(target)]
```
