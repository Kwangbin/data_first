```python
reserve_tb=pd.read_csv('reserve.csv', encoding='UTF-8')
reserve_tb
```

```python
# 합계값 계산
# hotel_id, total_price 추출하여 sum으로 매출 합계 계산

result_01 = reserve_tb.groupby(['hotel_id','people_num'])['total_price'].sum().reset_index()
result_01
```

```python
#total_price 를  price_sum 변경
result_01.rename(columns={'total_price': 'price_sum'},inplace=True)
result_01
```

```python
#최대 최소 대표값 찾기
#max, min, mean, media => total_price 대상
# agg 집역 처리

result_02 = reserve_tb.groupby('hotel_id').agg({'total_price':['max','min','mean','median',
                                                               lambda x: np.percentile(x,q=20)]}).reset_index()
result_02
```

```python
result_02.columns=['hotel_id','price_id','price_min','price_mean','price_median','price_20per']
result_02
```

```python
#분포계산
#분산 , 표준 편차

result_03= reserve_tb.groupby('hotel_id').agg({'total_price':['var','std']}).reset_index()
result_03
```

```python
result_03.columns=['hotel_id','var','std']
result_03
```

```python
#최빈값
# round로 반올림,mode 최빈값
reserve_tb['total_price'].round(-3).mode()
```

```python
#순위계산
#rank 사용
# 자료형이 문자열일때문 문제 발생 하니 숫자 , 날짜 형으로 변경

reserve_tb['reserve_datetime']= pd.to_datetime(reserve_tb['reserve_datetime'],format='%Y-%m-%d %H:%M:%S')
#log_no 열추가
# 고객별로 grup 및 순위계산 및 정렬
reserve_tb['log_no']=reserve_tb.groupby('customer_id')['reserve_datetime'].rank(ascending=True, method='first')
```

```python
#예약 횟수 계산
rsv_cnt_tb=reserve_tb.groupby('hotel_id').size().reset_index()
rsv_cnt_tb.columns=['hotel_id','rsv_cnt']
rsv_cnt_tb
```

```python
#cnt(예약횟수)를 기준으로 정렬

rsv_cnt_tb['rsv_cnt_rank']=rsv_cnt_tb['rsv_cnt'].rank(ascending=False,method='min')
rsv_cnt_tb
```
